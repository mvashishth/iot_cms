```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fftpack import fft
import math
import json
```

```python
def max_rms(z):
    z['X'], z['Y'],z['Z'], z['I'] = zip(*z['0'].map(lambda x: x.split(' ')))
    z['X']=z['X'].astype(float).fillna(0.0)
    z['Y']=z['Y'].astype(float).fillna(0.0)
    z['Z']=z['Z'].astype(float).fillna(0.0)
    
    maximum_x=float("{0:.2f}".format(np.max(abs(z['X']))))
    rms_x=float("{0:.2f}".format(np.sqrt(np.mean(np.square(z['X'])))))

    maximum_y=float("{0:.2f}".format(np.max(abs(z['Y']))))
    rms_y=float("{0:.2f}".format(np.sqrt(np.mean(np.square(z['Y'])))))
    
    maximum_z=float("{0:.2f}".format(np.max(abs(z['Z']))))
    rms_z=float("{0:.2f}".format(np.sqrt(np.mean(np.square(z['Z'])))))
    
    std_x= float("{0:.2f}".format(np.std(z['X'])))
    std_y= float("{0:.2f}".format(np.std(z['Y'])))
    std_z= float("{0:.2f}".format(np.std(z['Z'])))
    

    return maximum_x,rms_x,maximum_y,rms_y,maximum_z,rms_z, std_x, std_y, std_z
    
    
```
